```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# =========================
# 1. Google Drive Excel
# =========================
FILE_ID = "1Nog2oOuYR5ev8g34SwIEr10y72KvF8WF"
GID = "1478674873"

url = f"https://docs.google.com/spreadsheets/d/{FILE_ID}/export?format=xlsx&gid={GID}"

# =========================
# 2. Leer el Excel
# =========================
df = pd.read_excel(url)

# Asegurar formato correcto
df["Fecha"] = pd.to_datetime(df["Fecha"])
df = df.sort_values("Fecha")

# =========================
# 3. Estética del gráfico
# =========================
plt.rcParams.update({
    "font.family": "sans-serif",
    "font.size": 12,
    "axes.spines.top": False,
    "axes.spines.right": False
})

fig, ax = plt.subplots(figsize=(10, 5))

# =========================
# 4. Serie de tiempo
# =========================
ax.plot(
    df["Fecha"],
    df["Seguidores"],
    linewidth=2.8,
    alpha=0.9
)

#ax.set_title("Evolución de seguidores", fontsize=14, weight="bold")
ax.set_xlabel("Fecha")
ax.set_ylabel("Seguidores")

# Formato del eje X
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))

ax.grid(axis="y", linestyle="--", alpha=0.4)

plt.tight_layout()

# =========================
# 5. Guardar en alta resolución
# =========================
plt.savefig(
    "seguidores_tiempo.png",
    dpi=500,
    bbox_inches="tight"
)

plt.show()
```
